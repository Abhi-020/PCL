```python
import pandas as pd
import numpy as np
```

```python
np.random.randint(2, size=10)
```

```python
# making labels files

for i in range(5):
    
    labels = np.random.randint(2, size=100)
    nm = open(f'{i+1}.txt', 'w')
    for j in labels:
        print(j, file=nm)
    nm.close()
    
    
```

```python
# ref : https://www.kaggle.com/carlmcbrideellis/ensemble-methods-majority-voting-example

all_data = [pd.read_csv(f'{i+1}.txt', names=['label']) for i in range(5)]

votes = pd.concat(all_data, axis='columns')
votes
```

```python
predictions = votes.mode(axis='columns').to_numpy()

#f = open('final_ensemble.txt', 'w')
for i in predictions.flatten():
    print(i)#,file=f)
#f.close()
```
